```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

num_rows = 5000

customers = [f"CUST_{i}" for i in range(1, 501)]
products = [f"PROD_{i}" for i in range(1, 101)]
categories = ["Electronics", "Clothing", "Home", "Sports", "Beauty"]
cities = ["Bangalore", "Delhi", "Mumbai", "Chennai", "Hyderabad"]
payment_methods = ["UPI", "Credit Card", "Debit Card", "Cash on Delivery"]

data = []

start_date = datetime(2023, 1, 1)

for i in range(num_rows):
    order_id = f"ORD_{10000 + i}"
    customer_id = random.choice(customers)
    product_id = random.choice(products)
    category = random.choice(categories)
    
    price = round(random.uniform(100, 5000), 2)
    quantity = random.randint(1, 5)
    total_amount = price * quantity
    
    order_date = start_date + timedelta(days=random.randint(0, 730))
    payment_method = random.choice(payment_methods)
    city = random.choice(cities)
    
    data.append([
        order_id, customer_id, product_id, category,
        price, quantity, total_amount,
        order_date, payment_method, city
    ])

df = pd.DataFrame(data, columns=[
    "order_id", "customer_id", "product_id", "category",
    "price", "quantity", "total_amount",
    "order_date", "payment_method", "city"
])

df.to_csv("ecommerce.csv", index=False)

print("Dataset generated successfully!")
print(df.head())
```
